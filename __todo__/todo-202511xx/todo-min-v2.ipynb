```python
import omni.ui as ui

w = ui.Workspace.get_window("Viewport")  # or however you get it
flags = w.flags

print("raw flags:", flags)

# Map of all relevant flags you care about
flag_names = {
    "NONE": ui.WINDOW_FLAGS_NONE,
    "NO_TITLE_BAR": ui.WINDOW_FLAGS_NO_TITLE_BAR,
    "NO_RESIZE": ui.WINDOW_FLAGS_NO_RESIZE,
    "NO_MOVE": ui.WINDOW_FLAGS_NO_MOVE,
    "NO_SCROLLBAR": ui.WINDOW_FLAGS_NO_SCROLLBAR,
    "NO_SCROLL_WITH_MOUSE": ui.WINDOW_FLAGS_NO_SCROLL_WITH_MOUSE,
    "NO_COLLAPSE": ui.WINDOW_FLAGS_NO_COLLAPSE,
    # "ALWAYS_AUTO_RESIZE": ui.WINDOW_FLAGS_ALWAYS_AUTO_RESIZE,
    "NO_SAVED_SETTINGS": ui.WINDOW_FLAGS_NO_SAVED_SETTINGS,
    # "NO_INPUTS": ui.WINDOW_FLAGS_NO_INPUTS,
    "MENU_BAR": ui.WINDOW_FLAGS_MENU_BAR,
    # "HORIZONTAL_SCROLLBAR": ui.WINDOW_FLAGS_HORIZONTAL_SCROLLBAR,
    "NO_FOCUS_ON_APPEARING": ui.WINDOW_FLAGS_NO_FOCUS_ON_APPEARING,
    # "NO_BRING_TO_FRONT_ON_FOCUS": ui.WINDOW_FLAGS_NO_BRING_TO_FRONT_ON_FOCUS,
    # "ALWAYS_VERTICAL_SCROLLBAR": ui.WINDOW_FLAGS_ALWAYS_VERTICAL_SCROLLBAR,
    # "ALWAYS_HORIZONTAL_SCROLLBAR": ui.WINDOW_FLAGS_ALWAYS_HORIZONTAL_SCROLLBAR,
    # "NO_NAV_INPUT": ui.WINDOW_FLAGS_NO_NAV_INPUT,
    # "NO_NAV_FOCUS": ui.WINDOW_FLAGS_NO_NAV_FOCUS,
    # "UNSAVED_DOCUMENT": ui.WINDOW_FLAGS_UNSAVED_DOCUMENT,
    "NO_DOCKING": ui.WINDOW_FLAGS_NO_DOCKING,
    # add more if your Kit version has them
}

print("set flags:")
for name, value in flag_names.items():
    if flags & value:
        print(f"  - {name}")
```
